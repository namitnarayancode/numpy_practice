```python
import numpy as np
import matplotlib.pyplot as plt

# Create a sine wave
t = np.linspace(0, 1, 500)
signal = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*120*t)

# Fourier Transform
freq = np.fft.fft(signal)
freqs = np.fft.fftfreq(len(signal), d=t[1]-t[0])

# Plot spectrum
plt.plot(freqs[:250], np.abs(freq[:250]))
plt.show()
```

```python
student_data= np.dtype([
    ("name","U20"),
    ("marks","f4")
])
students = np.array([
    ('Alice', 85.5),
    ('Bob',  90.0),
    ('Charlie', 78.2)
],dtype=student_data)
np.savetxt("Numpy_5_6.txt",students,fmt="%s %f",delimiter=" ")
```

```python
student_dtype = np.dtype([("name","U20"),("marks","f4")])
loaded = np.genfromtxt("Numpy_5_6.txt", dtype=student_dtype, delimiter=" ")
print(loaded['name'])
print(loaded['marks'])
```
