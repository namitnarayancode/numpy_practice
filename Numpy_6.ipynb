```python
import numpy as np

# Define a structured dtype
dt = np.dtype([('name', 'U10'), ('age', 'i4'), ('height', 'f4')])

# Create array
x = np.array([('Alice', 25, 5.5),
              ('Bob',   30, 6.0),
              ('Cara',  22, 5.7)], dtype=dt)

print(x['name'])   # ['Alice' 'Bob' 'Cara']
print(x['age'])    # [25 30 22]
print(x['height']) # [5.5 6.  5.7]
print(x[['name','age']])
dt = np.dtype([('coords', 'f4', (3,))])  # field with 3 numbers
x = np.array([([1,2,3],), ([4,5,6],)], dtype=dt)
print(x['coords'])
# [[1. 2. 3.]
#  [4. 5. 6.]]
print(x.dtype.names)   # ('coords',)
```
