```python
arr= np.array([10, 20, 30, 40, 50])
mask = arr>25
print(mask)
print(arr[mask])
```

```python
arr = np.arange(1,11)
print(arr[(arr>3)&(arr<8)])
print(arr[(arr<4)|(arr>8)])
```

```python
mat= np.array([[11, 12, 13],
                [21, 22, 23],
                [31, 32, 33]])

print(mat[mat > 20 ])
```

```python
arr = np.arange(10)
print(arr)
mask = arr>5
print(arr[mask])
print(arr[[0,2,4]])
```

```python
mat = np.array([[1, 2, 3],
                [4, 5, 6]])

vec = np.array([10, 20, 30])

print(mat + vec)
```

```python
print(np.add(mat,10))
print(np.multiply(mat,5))
print(np.power(mat,3))
```

```python
angle= np.array([0,np.pi/2 , np.pi])
print(np.sin(angle))
print(np.cos(angle))
```

```python
angle= np.array([0,np.pi/2 , np.pi])
print(np.round(np.sin(angle)).astype(int))
print(np.round(np.cos(angle).astype(int)))
```

```python
a=np.array([0,np.pi/4,np.pi/2,np.pi])
b=np.power(np.sin(a),2)
c=np.power(np.cos(a),2)
print((b+c).astype(int)==1)
```

```python
print(np.round((np.tan([0,np.pi/6,np.pi/3])),2))
```

```python
np.round(np.exp([0,1,2]), 0)
```

```python
student_dtype=np.dtype([
    ("name","U10"),
    ("age", "i4"),
    ("score", "f4")
])
students = np.array([
    ('Alice', 21, 88.5),
    ('Bob',   19, 92.0),
    ('Charlie', 22, 79.5)   
],dtype=student_dtype)
print(students)
```
